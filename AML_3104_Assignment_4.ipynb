# Load the Dataset

```python
import tensorflow as tf

# Load the MNIST dataset from TensorFlow's built-in datasets
# Each image is a 28x28 grayscale image of a handwritten digit.

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train[1]
```

```python
import matplotlib.pyplot as plt

# display the sample training image

plt.imshow(x_train[1], cmap='gray')
```

# Preprocess the data

**Flatten the images first for machine learning model.**

ML 1D input vectors, in this step the images converted from 2D image to 1D vector (from 28 x 28 it becomes 784)

```python
# Flatten the images
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)

# Check the new shape
print("x_train_flat shape:", x_train_flat.shape)
print("x_test_flat shape:", x_test_flat.shape)
```

**Normalize Pixel Values**

Used in both ML and NN
Normalization improves learning speed and performance in all models

Each pixel is a grayscale value from 0 (black) to 255 (white).

I normalized the pixels by dividing it by 255

```python
# Normalize pixel values to range [0, 1]
x_train_norm = x_train_flat / 255.0
x_test_norm = x_test_flat / 255.0

# Optional: check min/max
print("Train min/max:", x_train_norm.min(), x_train_norm.max())
print("Test min/max:", x_test_norm.min(), x_test_norm.max())
```

After normalizing the pixels, we can see from the above output that the pixel ranges now between 0.0 -1.0 , where 0.0 = black, 1.0 = white and .5 is gray shades (in between)

# Train a Traditional ML Model  - Logistic Regression

Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create the model
ml_model  = LogisticRegression(max_iter=1000)  # 1000 iterations to ensure convergence

# Train the model
ml_model.fit(x_train_norm, y_train)

# Predict on test data
y_pred_lr = ml_model.predict(x_test_norm)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred_lr)

print("Test Accuracy (Logistic Regression):", accuracy)
```

```python
#Confusion Matrix

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred_lr)

# Display as heatmap
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(10))
disp.plot(cmap='Blues', xticks_rotation='vertical')

plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

# Train a Traditional ML Model  - kNN +  PCA

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import time

# 1. Apply PCA (reduce to 50 components)
pca = PCA(n_components=50)
x_train_pca = pca.fit_transform(x_train_norm)
x_test_pca = pca.transform(x_test_norm)

print(f"Original shape: {x_train_norm.shape}, Reduced shape: {x_train_pca.shape}")

# 2. Train k-NN on reduced data
knn_pca_model = KNeighborsClassifier(n_neighbors=3)

start_time = time.time()
knn_pca_model.fit(x_train_pca, y_train)
train_time = time.time() - start_time

# 3. Predict and evaluate
start_time = time.time()
y_pred_knn_pca = knn_pca_model.predict(x_test_pca)
test_time = time.time() - start_time

train_acc_knn_pca = knn_pca_model.score(x_train_pca, y_train)
test_acc_knn_pca = accuracy_score(y_test, y_pred_knn_pca)

print(f"\nk-NN (PCA-Reduced) Training Accuracy: {train_acc_knn_pca:.4f}")
print(f"k-NN (PCA-Reduced) Test Accuracy: {test_acc_knn_pca:.4f}")
print(f"Training Time: {train_time:.2f}s, Prediction Time: {test_time:.2f}s")
```

## Confusion Matrix for k-NN with PCA

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred_knn_pca)

# Create a display object
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(10))

# Plot
fig, ax = plt.subplots(figsize=(8, 8))
disp.plot(ax=ax, cmap='Blues')
plt.title("Confusion Matrix for k-NN + PCA")
plt.show()
```

## Save k-NN with PCA model

```python
from sklearn.pipeline import Pipeline
import pickle

# Combine PCA and k-NN into a pipeline
knn_pca_pipeline = Pipeline([
    ('pca', PCA(n_components=50)),
    ('knn', KNeighborsClassifier(n_neighbors=3))
])

# Fit the pipeline on normalized training data
knn_pca_pipeline.fit(x_train_norm, y_train)

# Save the model
with open("knn_model.pkl", "wb") as f:
    pickle.dump(knn_pca_pipeline, f)
```

# Build and Train the NN with 1 Hidden Layer

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the model
nn_model  = models.Sequential([
    layers.Input(shape=(784,)),              # Input layer (flattened image)
    layers.Dense(128, activation='relu'),    # Hidden layer
    layers.Dense(10, activation='softmax')   # Output layer (10 digits)
])

# Compile the model
nn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # Use sparse since labels are integers
    metrics=['accuracy']
)

# Train the model
nn_model.fit(x_train_norm, y_train, epochs=5, batch_size=32, validation_split=0.1)

# Evaluate on test data
test_loss, test_acc = nn_model.evaluate(x_test_norm, y_test)
print("Test Accuracy (Neural Network):", test_acc)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Predict with the NN
y_pred_nn = nn_model.predict(x_test_norm)
y_pred_labels = y_pred_nn.argmax(axis=1)  # Convert softmax outputs to label predictions

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(10))
disp.plot(cmap='Greens', xticks_rotation='vertical')
plt.title("Confusion Matrix - Neural Network")
plt.show()
```

# Trying NN with 2 Hidden Layers

```python
from tensorflow.keras import models, layers

# Build the model
nn_model_deep = models.Sequential([
    layers.Input(shape=(784,)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),      # New hidden layer
    layers.Dense(10, activation='softmax')
])

# Compile
nn_model_deep.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train
nn_model_deep.fit(
    x_train_norm, y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.1
)

# Evaluate
train_loss_deep, train_acc_deep = nn_model_deep.evaluate(x_train_norm, y_train, verbose=0)
test_loss_deep, test_acc_deep = nn_model_deep.evaluate(x_test_norm, y_test, verbose=0)

print(f"\n[2-Layer NN] Training Accuracy: {train_acc_deep:.4f}")
print(f"[2-Layer NN] Test Accuracy:     {test_acc_deep:.4f}")
```

## Adding Dropout

```python
from tensorflow.keras import models, layers

# Build the model with Dropout
nn_model_dropout = models.Sequential([
    layers.Input(shape=(784,)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),  # Dropout after 1st hidden layer
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),  # Dropout after 2nd hidden layer
    layers.Dense(10, activation='softmax')
])

# Compile
nn_model_dropout.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history_dropout = nn_model_dropout.fit(
    x_train_norm, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.1
)
```

```python
import matplotlib.pyplot as plt

# Extract metrics from training history
acc = history_dropout.history['accuracy']
val_acc = history_dropout.history['val_accuracy']
loss = history_dropout.history['loss']
val_loss = history_dropout.history['val_loss']
epochs = range(1, len(acc) + 1)

# Plot accuracy
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, acc, label='Training Accuracy')
plt.plot(epochs, val_acc, label='Validation Accuracy')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Predict class probabilities
y_pred_probs = nn_model_deep.predict(x_test_norm)

# Convert probabilities to class predictions
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred_classes)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(10))
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - NN with 2 Hidden Layers")
plt.show()
```

## Show Correct and Incorrect Predictions

```python
import numpy as np
import matplotlib.pyplot as plt

# Predict on test set
y_probs = nn_model_dropout.predict(x_test_norm)
y_preds = np.argmax(y_probs, axis=1)

# Identify correct and incorrect indices
correct_idxs = np.where(y_preds == y_test)[0]
incorrect_idxs = np.where(y_preds != y_test)[0]

# Function to plot images
def plot_predictions(indices, title, n=5):
    plt.figure(figsize=(12, 2))
    for i, idx in enumerate(indices[:n]):
        plt.subplot(1, n, i + 1)
        plt.imshow(x_test[idx], cmap='gray')
        plt.title(f"Pred: {y_preds[idx]}\nTrue: {y_test[idx]}")
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

# Show 5 correct predictions
plot_predictions(correct_idxs, "Correct Predictions")

# Show 5 incorrect predictions
plot_predictions(incorrect_idxs, "Incorrect Predictions")
```

```python
# Evaluate the trained Neural Network (with Dropout) on the training and test datasets.
# 'verbose=0' disables progress messages for cleaner output.

train_loss_dropout, train_acc_dropout = nn_model_dropout.evaluate(x_train_norm, y_train, verbose=0)
test_loss_dropout, test_acc_dropout = nn_model_dropout.evaluate(x_test_norm, y_test, verbose=0)
```
